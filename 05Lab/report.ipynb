```python
func = np.array([0.000000,\
				 0.021470,\
				 0.293050,\
				 0.494105,\
				 0.541341,\
				 0.516855,\
				 0.468617,\
				 0.416531,\
				 0.367879])
x = np.array([0.000,\
			  0.125,\
			  0.250,\
			  0.375,\
			  0.500,\
			  0.625,\
			  0.750,\
			  0.875,\
			  1.000])
```

## Метод трапеции

```python
def trapez(func, x):
	integral = 0
	for i in range(len(x) - 1):
		integral += (func[i]+func[i+1])*(x[i+1]-x[i])/2
	return integral
```

## Экстраполяция Ричардсона

```python
def rich(func, x):
	p = 2
	integral_h = trapez(func, x)
	
	x_2h = x[::2]
	func_2h = func[::2]

	integral_2h = trapez(func_2h, x_2h)

	integral = ((2**p)*integral_h - integral_2h)/(2**p - 1)
	return integral
```

## Метод Симпсона

```python
def simpson(func, x):
	integral = 0
	h = x[1] - x[0]
	odd = 0
	even = 0
	for i in range(len(x)-1):
		if(i % 2 == 0):
			odd += func[i]
		else:
			even += func[i]
	odd *= 2
	even *= 4
	integral = h*(func[0] + odd + even + func[-1])/3
	return integral
```
