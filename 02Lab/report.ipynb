## Введение норм и вспомогательных функций

### Модуль вектора

```python
def vec_abs(x, y):
	sum = 0
	for i in range(x.size):
		sum = sum + (x[i]-y[i])**2
	return np.sqrt(sum)
```

### Нормы векторов

```python
def normVec1(x):
	norm = 0.0
	for i in range(x.shape[0]):
		val = np.abs(x[i])
		if val > norm:
			norm = val
	return norm

def normVec2(x):
	norm = 0.0
	for i in range(x.shape[0]):
		norm += np.abs(x[i])
	return norm

def normVec3(x):
	norm_sq = np.dot(x, x)
	return np.sqrt(norm_sq)

def norm(x):
	return x/normVec3(x)
```

### Вычисление $\lambda_{max}$

```python
def calc_lambda(matrix):
	val = 0
	vec = np.arange(start=1, stop=size+1, step=1, dtype=double)
	for i in range(100):
		vec_start = vec
		vec = np.matmul(matrix,vec_start)
		lambda_vec = np.zeros(size)
		for i in range(size):
			lambda_vec[i] = vec[i]/vec_start[i]
		val = np.amax(lambda_vec)
	return val
```

### Нормы матриц

```python
def normMatr1(a):
	lines = a.shape[0]
	cols = a.shape[1]
	norm = 0.0
	for i in range(lines):
		val = 0.0
		for j in range(cols):
			val += np.abs(a[i,j])
		if val > norm:
			norm = val
	return val

def normMatr2(a):
	lines = a.shape[0]
	cols = a.shape[1]
	norm = 0.0
	for j in range(cols):
		val = 0.0
		for i in range(lines):
			val += np.abs(a[i,j])
		if val > norm:
			norm = val
	return val

def normMatr3(a):
	norm_sq = calc_lambda(np.matmul(a.transpose(), a))
	return np.sqrt(norm_sq)
```

## Методы, подобные методу Гаусса

### Проверка главных миноров матрицы

```python
def check_main_minors(matrix):
	for i in range(size):
		if(npl.det(matrix[:i, :i]) == 0):
			return False
	return True
```

### Метод LU

```python
def decompose_lu(matrix):
	n = matrix.shape[0]
	l = np.zeros([n, n])
	u = np.zeros([n, n])
	for i in range(n):
		l[i, i] = 1
	for i in range(n):
		for j in range(n):
			if i <= j:
				val = 0.0
				for k in range(i):
					val += l[i, k] * u[k, j]
				u[i, j] = matrix[i, j] - val
			else:
				val = 0.0
				for k in range(j):
					val += l[i, k] * u[k, j]
				l[i, j] = (matrix[i, j] - val)/u[j, j]
	return l, u
	
def forward_sub(matrix, column):
	n = matrix.shape[0]
	x = np.zeros(n, dtype=double)
	for i in range(n):
		tmp = cp.copy(column[i])
		for j in range(i):
			tmp -= matrix[i,j]*x[j]
		x[i] = tmp/matrix[i,i]
	return x
def back_sub(matrix, column):
	n = matrix.shape[0]
	x = np.zeros(n, dtype=double)
	for i in range(n-1, -1, -1):
		tmp = cp.copy(column[i])
		for j in range(i, n):
			tmp -= matrix[i,j]*x[j]
		x[i] = tmp/matrix[i,i]
	return x
def method_lu(matrix, column):
	#Decomposition
	L, U = decompose_lu(matrix)
	#Solve Ly=f
	y = forward_sub(L, cp.copy(column))
	#Solve Ux=y
	x = back_sub(U, y)
	return x
```

### Метод Гаусса с выбором главного элемента

```python
def method_main_elem(matrix, column):
	n = matrix.shape[0]
	gaussMatrix = cp.deepcopy(matrix)
	x = cp.deepcopy(column[:,0])
	swapedArr = np.zeros(shape=(0, 2))
	for k in range(n):
		matrCol = gaussMatrix[:, k]
		index = (np.where(np.abs(matrCol) == np.amax(np.abs(matrCol))))[0]
		if(index != k):
			np.append(swapedArr, [index, k])
			gaussMatrix[[index, k]] = gaussMatrix[[k, index]]
			x[index], x[k] = x[k], x[index]
		for i in range(k+1, n):
			coeff = gaussMatrix[i, k]/gaussMatrix[k,k]
			gaussMatrix[i, k] = 0
			for j in range(0, n):
				gaussMatrix[i, j] = gaussMatrix[i, j] - coeff*gaussMatrix[k, j]
			x[i] = x[i] - x[k]*coeff
	for k in range(n-1, -1, -1):
		for i in range(0, k):
			coeff = gaussMatrix[i, k]/gaussMatrix[k,k]
			x[i] = x[i] - x[k]*coeff
	for i in range(0,n):
		x[i] = x[i]/gaussMatrix[i,i]
	while swapedArr.size != 0:
		swapedItems = swapedArr[-1]
		first = swapedItems[0]
		second = swapedItems[1]
		x[first], x[second] = x[second], x[first]
		swapedArr.delete([-1])
	return x
```

### Выбор метода

```python
def method_choose(matrix, column):
	if(check_main_minors(matrix)):
		print("Выбран метод LU разложения")
		return method_lu(matrix, column)
	else:
		print("Выбран метод Гаусса с выбором главного элемента")
		return method_main_elem(matrix, column)
```

## Метод верхней релаксации

### Вычисление $\lambda$ для вычисления $\tau$

```python
lambda_max = calc_lambda(A)
lambda_min = 1/calc_lambda(npl.inv(A))
tau = 1/(lambda_max+lambda_min)
```

### Описание метода верхней релаксации

```python
def method_relax(matrix, column, tau):
	n = matrix.shape[0]
	matrixB = np.eye(n) - np.dot(matrix,tau)
	columnF = np.dot(column,tau)
	x = np.array([1]*n, dtype=double)
	while True:
		x_start = x
		x = np.matmul(matrixB, x_start)+columnF
		if(vec_abs(x_start, x) < 1e-5):
			break
	return x
```

## Методы для определения невязок и числа обусловленностей

```python
def discrepancy(matrix, column, solution, func):
	return func(np.matmul(matrix, solution) - column[:,0])

def mu_calc(matrix, func):
	return func(matrix) * func(npl.inv(matrix))
```
